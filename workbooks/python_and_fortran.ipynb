```python
from numpy import *
from matplotlib.pyplot import *

# http://arogozhnikov.github.io/2015/11/29/using-fortran-from-python.html
%load_ext fortranmagic
```

```python
%%fortran
subroutine my_function(x, y, z)
    real, intent(in) :: x(:), y(:)
    real, intent(out) :: z(size(x))
    ! using vector operations
    z(:) = max(x(:), y(:))
end subroutine
```

```python
x = random.normal(size=100)
y = random.normal(size=100)
z = my_function(x, y)
```

```python
plot(x)
plot(y)
plot(z)
```
